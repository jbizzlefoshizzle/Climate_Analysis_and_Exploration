# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
## Last date on csv
latest_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
## Extract string (remove parentheses)
latest_date = list(np.ravel(latest_date))[0]
## Convert to datetime to make life easier
latest_date = dt.datetime.strptime(latest_date, '%Y-%m-%d')

# Calculate the date 1 year ago from the last data point in the database
latest_year = int(dt.datetime.strftime(latest_date, '%Y'))
latest_month = int(dt.datetime.strftime(latest_date, '%m'))
latest_day = int(dt.datetime.strftime(latest_date, '%d'))
year_before = dt.date(latest_year, latest_month, latest_day) - dt.timedelta(days=365)

# Perform a query to retrieve the data and precipitation scores
rain_data = session.query(Measurement.date, Measurement.prcp).\
filter(Measurement.date > year_before).\
order_by(Measurement.date).\
all()

# Save the query results as a Pandas DataFrame and set the index to the date column
rain_table = pd.DataFrame(rain_data)
rain_table = rain_table.set_index('date')

# Sort the dataframe by date
rain_df = rain_table.sort_index(ascending=True)

# Use Pandas Plotting with Matplotlib to plot the data
rain_df.plot(title="Precipitation\n" + "8/23/16 - 8/23/17", figsize=(16,8))
```

```python
# Design a query to show how many stations are available in this dataset?
station_count = session.query(Station).count()
print(f'Station Count: {station_count}')
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
active_stations = session.query(Measurement.station, func.count(Measurement.station)).\
group_by(Measurement.station).\
order_by(func.count(Measurement.station).\
         desc()).all()
active_stations
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?

most_active_num = active_stations[0][0]
station_name = session.query(Station.name).filter_by(station = most_active_num)
most_active_station = station_name[0][0]
print(f'Most Active Station: {most_active_num} - {most_active_station}')

lowest_temp = session.query(Measurement.tobs).filter(Measurement.station == most_active_num).order_by(Measurement.tobs.asc()).first()
print(f'Lowest Temperature Recorded Here: {lowest_temp}')

highest_temp = session.query(Measurement.tobs).filter(Measurement.station == most_active_num).order_by(Measurement.tobs.desc()).first()
print(f'Highest Temperature Recorded Here: {highest_temp}')

avg_temp = session.query(func.avg(Measurement.tobs)).filter(Measurement.station == most_active_num)
avg_temp = '{0:.6}'.format(avg_temp[0][0])
print(f'Average Temperature Recorded Here: {avg_temp}')
```

```python
# Choose the station with the highest number of temperature observations.
## (Isn't this still USC00519281?)
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
temp_data = session.query(Measurement.date, Measurement.tobs).filter(Measurement.date > year_before).filter(Measurement.station == most_active_num).order_by(Measurement.date).all()
temp_temp = pd.DataFrame(temp_data)
bweh = temp_temp.set_index('date')
temp_df = bweh.sort_index(ascending=True)
temp_df
temp_df.plot(kind='hist', bins=12, figsize=(16,8), alpha = 0.5)
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.

trip_start = '2017-06-30'
trip_end = '2017-07-08'
trip_temps = calc_temps(trip_start, trip_end)
trip_temps
```

```python
# Plot the results from your previous query as a bar chart.
trip_min = trip_temps[0][0]
trip_avg = trip_temps[0][1]
trip_max = trip_temps[0][2]

# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
error = trip_max - trip_min

# Use "Trip Avg Temp" as your Title
vacay = '2019-06-30 to\n 2019-07-08'
# Use the average temperature for the y value
plt.bar(vacay, trip_avg, yerr=error, color = 'green', alpha = .5)
plt.ylim(0,100)
plt.xlim(-2,2)
plt.title('Trip Avg Temp')
plt.ylabel('Temp (F)')
plt.show()
```

```python
data_stuff = [Station.station,
             Station.name,
             func.sum(Measurement.prcp),
             Station.latitude,
             Station.longitude,
             Station.elevation]

total_rain = session.query(*data_stuff).filter(Station.station == Measurement.station).\
filter(Measurement.date >= trip_start).\
filter(Measurement.date <= trip_end).\
group_by(Station.station).\
order_by(func.sum(Measurement.prcp).\
         desc()).all()

total_rain_table = pd.DataFrame(np.array(total_rain))

total_rain_df = total_rain_table.rename(columns={0: "Station", 1: "Location", 2: "Total Rainfall", 3: "Latitude", 4: "Longitude", 5: "Elevation"})
total_rain_df
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
